```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid
```

```python
#Choosing Early Universe version of the Hubble Constant converted in to /Gyr:

H_0=0.06847
```

```python
#Case 1:

def universe_1(u):
  return (H_0**-1)*(u**-1)**(-1/2)

a_1 = np.linspace(0.01, 1, 500)
integrand_1 = universe_1(a_1)

uni_1 = cumulative_trapezoid(integrand_1, a_1, initial=0)

def universe_2(u):
  return (H_0**-1)*(0.3*u**-1 + 0.7*u**2)**(-1/2)

uni_2 = cumulative_trapezoid(universe_2(a_1), a_1, initial=0)

def universe_3(u):
  return (H_0**-1)*(0.3*u**-1 + 0.7*u**2.6)**(-1/2)

uni_3 = cumulative_trapezoid(universe_3(a_1), a_1, initial=0)

def universe_4(u):
  return (H_0**-1)*(0.3*u**-1 + 0.7*u**1.4)**(-1/2)

uni_4 = cumulative_trapezoid(universe_4(a_1), a_1, initial=0)

def universe_5_plus(u):
  return ((H_0+0.005109)**-1)*(0.3*u**-1 + 0.7*u**2)**(-1/2)

uni_5_plus = cumulative_trapezoid(universe_5_plus(a_1), a_1, initial=0)

def universe_5_minus(u):
  return ((H_0-0.005109)**-1)*(0.3*u**-1 + 0.7*u**2)**(-1/2)

uni_5_minus = cumulative_trapezoid(universe_5_minus(a_1), a_1, initial=0)

plt.figure()
plt.plot(a_1, uni_1, label="Case 1", color="green")
plt.plot(a_1, uni_2, label="Case 2", color="red")
plt.plot(a_1, uni_3, label="Case 3", color="blue")
plt.plot(a_1, uni_4, label="Case 4", color="orange")
plt.plot(a_1, uni_5_plus, label="Case 5+", color="purple")
plt.plot(a_1, uni_5_minus, label="Case 5-", color="black")
plt.legend()
plt.xlabel("Scaling factor - a")
plt.ylabel("t(a)(Gyr)")
plt.title("Universe Plot")
plt.show()
```
