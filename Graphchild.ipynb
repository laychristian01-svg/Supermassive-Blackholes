```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid
```

```python
#Choosing Early Universe version of the Hubble Constant converted in to /Gyr:

H_0=0.06847
```

```python
#Case 1:

def universe_1(u):
  return (H_0**-1)*(u**-1)**(-1/2)

a_1 = np.linspace(0.01, 1, 500)
integrand_1 = universe_1(a_1)

uni_1 = cumulative_trapezoid(integrand_1, a_1, initial=0)



plt.figure()
plt.plot(a_1, uni_1, label="Case 1", color="green")
plt.xlabel("a")
plt.ylabel("t(a)(Gyr)")
plt.title("Universe Plot")
plt.show()
```
